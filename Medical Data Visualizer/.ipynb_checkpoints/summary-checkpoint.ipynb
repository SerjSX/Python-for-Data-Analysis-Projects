## Importing neeeded packages

We need the following packages for this project: pandas, seaborn, matplotlib and numpy

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Loading the CSV file

We are going to use .read_csv provided by pandas to load the CSV file that includes the medical examination information. The first 5 rows of the dataset is loaded for display.

```python
df = pd.read_csv('medical_examination.csv')
df.head(5)
```

## Creating Overweight column

First, we will create a new column named `overweight` to determine if an individual is overweight or not. We are going to determine this factor by calculating the `BMI`.

`BMI`: we divide the `weight` that is in kilograms by the square of their `height` that is in meters. 
- We use `/100` to convert the height to kilograms, since height is in centemers.
- We use `**2` to find the square of the height after the division by 100.

```python
df['BMI'] = df['weight'] / (df['height'] / 100)**2
```

- If the BMI's value is > 25 then the person is overweight, we assign the value 1 to the new column we created named "overweight".
- If it's not greater than 25, then we assign the value 0 to the new column stating that the person is not overweight.
- We delete the column named BMI since we do not need it anymore.

```python
df['overweight'] = df['BMI'].apply(lambda x: 1 if x > 25 else 0)
del df['BMI']
```

## Normalizing the data

We are going to normalize the data by assigning the value `0` to always good, and the value `1` to always bad. If the value of `cholesterol` or `gluc` is 1, set the value to `0`. If the value is more than `1`, set the value to `1`.

- lambda x: is used to apply the condition. If it's equal to 1, then apply 0. If not then 1.
- We are updating the values of the columns `cholesterol` and `gluc` only.

```python
df['cholesterol'] = df['cholesterol'].apply(lambda x: 0 if x == 1 else 1)
df['gluc'] = df['gluc'].apply(lambda x: 0 if x == 1 else 1)
```

## Draw Categorical Plot 

Next we will start to draw 2 charts that show the counts of categories we have for both cardio `0` and cardio `1`. 
- Cardio 0: presence of cardiovascular disease
- Cardio 1: absence of cardiovascular disease

The chart will show the values of `cholesterol`, `gluc`, `smoke`, `alco`, `active`, and `overweight`. 

Additionally, since each of these category use `0` or `1` to indicate the absence/prescence of having XYZ, the categories within the two graphs will have 2 bars. Once for 0, and one for 1.

Here is what `0`, `1` means for each of the categories:
- **Cholesterol:** good cholesterol, bad cholesterol
- **Gluc:** good glucose, bad glucose
- **Smoke:** no smoking, smoking
- **Alco:** does not drink alcohol, drinks alcohol
- **Active:** is not active, is active (physical activity)
- **Overweight:** is not overweight, is overweight

---
### Grouping and Reformatting the Data 
1. We will only use the values of the columns cholesterol, gluc, smoke, alco, active, and overweight. In order to filter to those columns only, we are adding `value_vars` in `pd.melt()`.
2. We specify df as the dataframe to be used by `pd.melt()`.
3. `id_vars` states by which column should we apply the id as. We are going to do this based off the values in the column `cardio`, so this automatically inserts the cardio column as a separate column.

The `value_vars` will be stored as a separate column named `variable`, and the `value` column will store each value of the items in the `variable` column.

So if we have the following data:

    cardio | gluc | smoke
    
    1      | 1    | 1
    
    0      | 0    | 1

The `pd.melt()` function with the paremeters we added (of course we don't have cholesterol, smoke, and a couple of others in the above example. But we can easily fix that by removing the columns we don't need from the `value_vars` paremeter's list) will result the following:

    cardio | variable | value
    
    1      | gluc     | 1
    
    1      | smoke    | 1
    
    0      | gluc     | 0
    
    0      | gluc     | 1

Which is exactly what we need. Here is the full code for the operation the medical examination data:

```python
df_cat = pd.melt(df, id_vars='cardio', value_vars=["cardio", "cholesterol", "gluc", "smoke", "alco", "active", "overweight"])
print(df_cat)
```

### Converting, sorting and creating the plot

#### Converting and Sorting 

We are going to sort the values by the variable names (carido, cholesterol, gluc...), so we would see the chart in alphabetical order.

```python
df_cat = pd.DataFrame(df_cat).sort_values('variable')
```

#### Creating the plot 

Now we will create the plot. As we mentioned earlier, the plot will display the total count of each category we have, whether 0 or 1 for the category, in both cardio 0 and 1. 

Hence, we will have **two charts** showing the total count of two bars for each of the categories, one for 0 and one for 1. 

We are going to use `sns.catplot()` provided by Seaborn stored in the `fig` variable. Here is what each paremeter is used for:
1. `data`: points to the dataset (dataframe) we are going to visualize.
2. `kind`: what the plot/chart should be of type, in this case we are going to use `count` because we need to count how many values we have in each category. Here are the options for this paremeter: strip plot, swarm plot, box plot, violin plot, boxen plot, point plot, bar plot and **count plot**,
3. `x`: what to display on the x-axis, in this case we wnant to display the `variable` values which contains the category names cholesterol, gluc, smoke...
4. `hue`: what to separate the bars as for each category, and in this case we want to separate them by the values in the `value` column which holds either `0` or `1` for each category.
5. `col`: what to use as a separator column, which indicates when to use another chart bassed off the values of the column. Here we will use the column `cardio`, which contains 0 or 1. So we will have 2 charts: one for the count values of categories of cardio `0`, and another for `1`.

Here is the final code:

```python
sns.set_style("whitegrid")
fig = sns.catplot(data=df_cat, kind='count', x='variable', hue='value', col='cardio').set_ylabels('total').fig
```

## Heat Map
### Cleaning Data

Conditions for incorrect data to clean:
- diastolic pressure `ap_lo` is higher than systolic pressure `ap_hi`
- `height` is less than the 2.5th percentile
- `height` is more than the 97.5th percentile
- `weight` is less than the 2.5th percentile
- `weight` is more than the 97.5th percentile

We use `.quantile()` to calculate the Xth percentile. We put the percentage in the parenthesis.

```python
df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
            (df['height'] >= df['height'].quantile(0.025)) &
            (df['height'] <= df['height'].quantile(0.975)) &
            (df['weight'] >= df['weight'].quantile(0.025)) &
            (df['weight'] <= df['weight'].quantile(0.975))]
```

### Correlation Matrix

We are going to calculate the correlation matrix for each of the values in and store it in the `corr` variable. 

We are going to use `.corr()` to perform this calculation.

```python
corr = df_heat.corr()
```

### Mask for Upper Triangle

We are going to create a mask for the upper triangle of the `corr` dataframe in order to determine the format we want of the heatmap.

We will be replicating the dimensions and matrix of the data in the `corr` variable by using `.ones_like`. This way the default values won't be used.

We are going to use `np.triu()` in order to create the upper triangle.

```python
mask = np.triu(np.ones_like(corr))
print(mask)
```

As you can see, the ones look like the triangle, which is precisely what we want. Later when we create the heatmap, we will add a paremeter named `mask`. We would pass the `mask` variable we created above, and this paremeter will only place values in the places that are NOT 1. In this case, only the 0s that are shown above will be actually filled in the heatmap.

### Preparing Figure

We are preparing the figure and the axis with an appropriate size that makes the heatmap easily viewable afterwards. Now if we draw the heatmap it will be added on this subplot that we created.

- `fig` shows information about the subplot, such as the size and what there is in it.
- `ax` displays the actual graph/plot.

```python
fig,ax=plt.subplots(figsize=(14,10))
print(fig)
print(ax)
```

### Heatmap

Now we are going to create the actual heatmap with the proper configuration.
1. The `corr` is provided as the first paremeter's data in order to specify that we want to visualize the values of this dataset.
2. We are going to specify the `mask` paremeter by assinging the `mask` variable that we created above (they have the same name). This will ignore the places that are ones, and will only display data in the places where is 0.
3. `annot=True` is specified in order to show the numbers in each of the cells in the heatmap, or else there won't be numbers to specify the values. This makes it easier for us to see.
4. Last, we are specifying a format to be used. `fmt='.1f'` allows us to see only one digit after the decimal place.

```python
sns.heatmap(corr, mask=mask, annot=True, fmt='.1f')
```

The brighter the color, the stronger the correlation. 
- The strongest correlation exists between weight and overweight, where it is 0.7. This indicates that as weight increases, the person becomes overweight at a higher rate.
- The mid-strong correlation exists between height and sex, which can be true sometimes. The sex of the person can sometimes affect the height.
- Minimal linear trend exists between smoking and sex, cardio and diastolic blood pressure, glucose and cholesterol, and alcohol and smoking.
- No linear trend exists between the rest of the dark colors.
- Weak correlation between overweight and height. This may mean that as one increases, the other decreases.
